# Análisis de datos - Desafío Alura Store

En este cuaderno vamos a analizar el rendimiento de las 4 tiendas para ayudar al señor Juan a decidir cuál debería vender. Usaremos los siguientes criterios:

1. Facturación total de cada tienda
2. Categorías más populares
3. Promedio de evaluación
4. Productos más y menos vendidos
5. Costo promedio de envío


```python
# Importación de librerías y carga de datos
import pandas as pd
import matplotlib.pyplot as plt

# URLs de los archivos CSV en GitHub
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargamos los datos

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


```

## 1. Facturación total por tienda
Vamos a calcular cuánto facturó cada tienda sumando el total de los precios de sus productos vendidos.


```python
facturacion_total = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

# Visualización
plt.bar(facturacion_total.keys(), facturacion_total.values(), color='plum')
plt.title('Facturación Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación Total')
plt.show()
```

## 2. Categorías más populares por tienda
Analizamos cuáles son las categorías de productos más vendidas en cada tienda.


```python
# Categorías populares por tienda
# Tienda 1
categorias_t1 = tienda1["Categoría del Producto"].value_counts()
categorias_t1.plot(kind="bar", title="Categorías más populares - Tienda 1", color='#00f0f0')
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 2
categorias_t2 = tienda2["Categoría del Producto"].value_counts()
categorias_t2.plot(kind="bar", title="Categorías más populares - Tienda 2", color='lightgreen')
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 3
categorias_t3 = tienda3["Categoría del Producto"].value_counts()
categorias_t3.plot(kind="bar", title="Categorías más populares - Tienda 3", color='orange')
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tienda 4
categorias_t4 = tienda4["Categoría del Producto"].value_counts()
categorias_t4.plot(kind="bar", title="Categorías más populares - Tienda 4", color='skyblue')
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##  Promedio de evaluación por tienda


```python
# Calcular el promedio de calificación por tienda
promedios = {
    "Tienda 1": tienda1["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

print(promedios)
```

```python
# Graficar los promedios
import matplotlib.pyplot as plt

# Crear figura
plt.figure(figsize=(8, 5))

# Graficar barras
plt.bar(promedios.keys(), promedios.values(), color='#eab676')

# Etiquetas y título
plt.title("Promedio de Calificación por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")

# Mostrar gráfico
plt.show()
```

## 3. Promedio de calificación por tienda

```python
# Promedio de calificación por tienda
promedio_calificaciones = {
    "Tienda 1": tienda1["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

# Mostramos el resultado
print("Promedio de calificación por tienda:")
for tienda, promedio in promedio_calificaciones.items():
    print(f"{tienda}: {promedio:.2f}")
```

```python
# Gráfico del promedio de calificación
# Importamos librerías
import matplotlib.pyplot as plt

# Creamos la gráfica de barras
plt.figure(figsize=(8, 5))
plt.bar(promedio_calificaciones.keys(), promedio_calificaciones.values(), color="purple")
plt.title("Promedio de Calificación por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)  # Las calificaciones van de 0 a 5
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

## 4. Productos más y menos vendidos por tienda

```python
# Paso 4: Productos más y menos vendidos por tienda

# Función para mostrar el producto más y menos vendido de cada tienda
def productos_mas_menos_vendidos(nombre_tienda, df_tienda):
    conteo_productos = df_tienda["Producto"].value_counts()
    mas_vendido = conteo_productos.idxmax()
    menos_vendido = conteo_productos.idxmin()
    print(f"{nombre_tienda}:")
    print(f"🔺 Producto más vendido: {mas_vendido} ({conteo_productos.max()} ventas)")
    print(f"🔻 Producto menos vendido: {menos_vendido} ({conteo_productos.min()} venta)\n")

# Ejecutamos la función para cada tienda
productos_mas_menos_vendidos("Tienda 1", tienda1)
productos_mas_menos_vendidos("Tienda 2", tienda2)
productos_mas_menos_vendidos("Tienda 3", tienda3)
productos_mas_menos_vendidos("Tienda 4", tienda4)
```

## 5. Costo promedio de envío por tienda

```python
# Paso 5: Costo promedio de envío por tienda

# Creamos un diccionario con los promedios de cada tienda
envio_promedio = {
    "Tienda 1": tienda1["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

# Mostramos los resultados
print("Costo promedio de envío por tienda:")
for tienda, costo in envio_promedio.items():
    print(f"{tienda}: ${costo:,.2f}")
```

## Gráfico del costo de envío por tienda

```python
# Gráfico de costo promedio de envío
sns.barplot(x=list(envio_promedio.keys()), y=list(envio_promedio.values()), palette="pastel")
plt.title("Costo promedio de envío por tienda")
plt.ylabel("Costo en pesos")
plt.xlabel("Tienda")
plt.xticks(rotation=45)
plt.show()
```

##  Resumen final

```python
# el DataFrame con las métricas por tienda)
# Creamos un resumen de indicadores clave por tienda
# Calcular promedios de calificación
promedios_calificacion = {
    "Tienda 1": tienda1["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

# Calcular costos promedio de envío
costos_envio = {
    "Tienda 1": tienda1["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

# Crear resumen
resumen = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturación Total": list(facturacion_total.values()),
    "Promedio Evaluación": list(promedios_calificacion.values()),
    "Costo Promedio Envío": list(costos_envio.values())
})

# Mostrar resumen ordenado
resumen.sort_values(by="Facturación Total", ascending=False)
```

## Visualización del resumen (gráfico de barras múltiples)

```python
# Gráfico 1: Facturación Total
plt.figure(figsize=(8, 4))
plt.bar(resumen["Tienda"], resumen["Facturación Total"], color="skyblue")
plt.title("Facturación Total por Tienda")
plt.ylabel("Facturación")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()

# Gráfico 2: Promedio Evaluación
plt.figure(figsize=(8, 4))
plt.bar(resumen["Tienda"], resumen["Promedio Evaluación"], color="lightgreen")
plt.title("Promedio de Evaluación por Tienda")
plt.ylabel("Evaluación Promedio")
plt.xlabel("Tienda")
plt.ylim(0, 5)  # Evaluaciones de 1 a 5
plt.tight_layout()
plt.show()

# Gráfico 3: Costo Promedio de Envío
plt.figure(figsize=(8, 4))
plt.bar(resumen["Tienda"], resumen["Costo Promedio Envío"], color="salmon")
plt.title("Costo Promedio de Envío por Tienda")
plt.ylabel("Costo Promedio")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```
